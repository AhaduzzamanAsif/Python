```python
import copy
import random
currState=[[0,0,0,0,0,0],[0,1,2,4,0,0],[0,5,6,3,7,0],[0,9,10,11,8,0],[0,13,14,15,12,0],[0,0,0,0,0,0]]
goalState=[[0,0,0,0,0,0],[0,1,2,3,4,0],[0,5,6,7,8,0],[0,9,10,11,12,0],[0,13,14,15,0,0],[0,0,0,0,0,0]]
move=[[0,1],[0,-1],[-1,0],[1,0]]

pos=[1,4]

def position(x):
  i=0
  for rows in goalState:
    if x in rows:
      p=[i,rows.index(x)]
      return p
    i=i+1
def hCost(temp):
  cost=0
  for i in range(1,5):
    for j in range(1,5):
      if temp[i][j]!=0:
        p=position(temp[i][j])
        cost+=abs(p[0]-i)+abs(p[1]-j)
  return cost

def setCurrState(index,pos):
  currState[pos[0]][pos[1]],currState[pos[0]+move[index][0]][pos[1]+move[index][1]]=currState[pos[0]+move[index][0]][pos[1]+move[index][1]],currState[pos[0]][pos[1]]

def getIndex(cost,mcost):
  index=[]
  for i in range(len(cost)):
    if cost[i]==mcost:
      index.append(i)
  indx = random.choice(index)
  return indx

def printMatrix(matrix):
  for i in range(1,5):
    for j in range(1,5):
      print(f"|{str(matrix[i][j]).rjust(2)}", end=' ')
      print('|',end='')
    print()
print('Current State:')
printMatrix(currState)
mcost=999
while(mcost!=0):
  cost=[999 for i in range(6)]
  for i in range(4):
    temp=copy.deepcopy(currState)
    if(temp[pos[0]+move[i][0]][pos[1]+move[i][1]]!=0):
      temp[pos[0]][pos[1]],temp[pos[0]+move[i][0]][pos[1]+move[i][1]]=temp[pos[0]+move[i][0]][pos[1]+move[i][1]],temp[pos[0]][pos[1]]
      print()
      if i==0: print('Right Shift:')
      elif i==1: print('Left Shift:')
      elif i==2: print('Up Shift:')
      elif i==3: print('Down Shift:')
      printMatrix(temp)
      cost[i]=hCost(temp)
      print('Heuristic Cost: ',cost[i])
  mcost=min(cost)
  index=getIndex(cost,mcost)
  setCurrState(index,pos)
  print()
  print('New Current State:')
  printMatrix(currState)
  print('Heuristic Cost: ',mcost)
  pos[0],pos[1]=pos[0]+move[index][0],pos[1]+move[index][1]
print('........Reaching the goal state.........')
```
